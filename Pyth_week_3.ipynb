*Q1. Lets say we have a list L1 = [1,2,2,3,2,3,4,5] now identify the most repetitive element in the list?*

```python
num_counter={}
L1= [1,2,2,3,2,3,4,5]
for l in L1:
  if l in num_counter:
    num_counter[l]+=1
  else:
    num_counter[l]=1

# h=max(num_counter,key=num_counter.get)
# print(h)

temp = 0
for k,v in num_counter.items():
  if v>temp:
    temp = v
    final_result= k
print(final_result,"has", temp, "entries")
```

```python
L1 = [1,2,2,3,2,3,4,5]

most_frequent_element = None
max_frequency = 0

# Get unique elements without using a set (which is also a type of hash table/dictionary behind the scenes)
# A simpler way if we don't want to use sets is to create a new list of unique elements
unique_elements = []
for item in L1:
    if item not in unique_elements:
        unique_elements.append(item)

for element in unique_elements:
    current_frequency = 0
    for item in L1:
        if item == element:
            current_frequency += 1

    if current_frequency > max_frequency:
        max_frequency = current_frequency
        most_frequent_element = element

print(f"The most repetitive element is {most_frequent_element} with a frequency of {max_frequency}.")
```
